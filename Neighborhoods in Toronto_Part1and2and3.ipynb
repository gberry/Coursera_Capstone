```python
# Advice on how to webscrape Wikipedia came from: https://simpleanalytical.com/how-to-web-scrape-wikipedia-python-urllib-beautiful-soup-pandas
# Note: I believe the wikipedia changed some from when the assignmen was made bacause there are no duplicate PostalCode rows (that combine has 
# already been done on the wikipedia page, and there are no rows with a borough but no neighborhood)
import requests
website_url = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text

#Use BeautifulSoup to find the table in the wikipedia page
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
My_table = soup.find('table',{'class':'wikitable'})

```

```python
# Extract all the column values for each row in the table and append to appropriate list
A=[]
B=[]
C=[]
for row in My_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        Aval = str(cells[0].find(text=True)).strip()
        A.append(Aval)
        Bval = str(cells[1].find(text=True)).strip()
        B.append(Bval)
        Cval = str(cells[2].find(text=True)).strip()
        C.append(Cval)
        
# Build pandas dataframe from lists         
import pandas as pd
df=pd.DataFrame(A,columns=['PostalCode'])
df['Borough']=B
df['Neighborhood']=C
```

```python
# Clean the DataFrame according to the rules set in the exercise
# 1. drop rows where Borough is "Not assigned"
newdf=df.query('Borough != "Not assigned"')
# 2. aggregate all neighborhoods for the same combination of PostalCode and Neighborhood
mydf = newdf.groupby(['PostalCode','Borough'], as_index=False)['Neighborhood'].agg(lambda x: x)
#Convert / to , : I don't think this was in the original assignment but has been added to 
#account for changes to the wikipedia page
mydf["Neighborhood"] = mydf["Neighborhood"].str.replace(" /",",")
mydf
```

```python
results = requests.get(url).json()
results
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```
